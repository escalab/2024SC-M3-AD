```python
import seaborn as sns
import pandas as pd
from matplotlib.ticker import FuncFormatter
import matplotlib.pylab as plt

mrf_result = pd.read_csv("mrf_result.csv")
custom = {"axes.edgecolor": "black", "grid.linestyle": "dashed", "grid.color": "black"}
sns.set_style("whitegrid", rc = custom)
g = mrf_result.set_index('NATOMS',).plot(kind='bar', 
                    stacked=True,width=0.65, 
                    color=sns.color_palette('crest'),
                    fontsize=20)
sns.set_theme()
plt.ylim(0, 1.3)
g.set_xlabel('Number of atoms',fontsize=27)
g.set_ylabel('Speedup',fontsize=28)
g.get_legend().remove()
g.grid(axis='x')
g.set_xticklabels(mrf_result['NATOMS'],rotation=90,fontsize=22)
g.set_yticklabels(mrf_result['speedup'],fontsize=22)
# Format yticks with .1f
formatter = FuncFormatter(lambda x, _: f'{x:.1f}')
g.yaxis.set_major_formatter(formatter)

fig = g.figure
fig.set_figwidth(8)
fig.set_figheight(2.5)
plt.show()
fig.savefig('/nfshome/yuz057/2023MICRO-MPCMXU/Figures/mrf.pdf',bbox_inches='tight')
```
