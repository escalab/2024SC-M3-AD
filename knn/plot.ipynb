```python
import matplotlib.pyplot as plt
import seaborn as sns

# sns.set_theme()

fig, axs = plt.subplots(2, 2)
df = pd.read_csv('result_a100.csv')
df_05k = df[df['dim'] == 512]
df_1k = df[df['dim'] == 1024]
df_2k = df[df['dim'] == 2048]
df_4k = df[df['dim'] == 4096]
df_05k = df_05k.pivot(index='query', columns='ref', values='speedup')
df_1k = df_1k.pivot(index='query', columns='ref', values='speedup')
df_2k = df_2k.pivot(index='query', columns='ref', values='speedup')
df_4k = df_4k.pivot(index='query', columns='ref', values='speedup')

# plot in the first subplot
sns.heatmap(data=df_05k, ax=axs[0, 0], cmap='crest',annot=True, fmt=".1f", cbar=False, linewidths=.5, annot_kws={"fontsize": 12})
# g = sns.heatmap(df, cmap='crest', annot=True, fmt=".1f", cbar=False, linewidths=.5)
axs[0, 0].set_title('Dim = 512', size=15)
axs[0, 0].set_xticks([])
axs[0, 0].set_xlabel('')
axs[0, 0].set_ylabel('Query', size=18)
axs[0, 0].tick_params(axis='y', which='both', length=0, labelsize=13)

# plot in the second subplot
sns.heatmap(data=df_1k, ax=axs[0, 1], cmap='crest',annot=True, fmt=".1f", cbar=False, linewidths=.5, annot_kws={"fontsize": 12})
axs[0, 1].set_title('Dim = 1024', size=15)
axs[0, 1].set_xticks([])
axs[0, 1].set_yticks([])
axs[0, 1].set_xlabel('')
axs[0, 1].set_ylabel('')

# plot in the third subplot
sns.heatmap(data=df_2k, ax=axs[1, 0], cmap='crest',annot=True, fmt=".1f", cbar=False, linewidths=.5, annot_kws={"fontsize": 12})
axs[1, 0].set_title('Dim = 2048', size=15)
axs[1, 0].set_ylabel('Query', size=18)
axs[1, 0].set_xlabel('Reference', size=18)
axs[1, 0].tick_params(axis='x', which='both', length=0, labelsize=13)
axs[1, 0].tick_params(axis='y', which='both', length=0, labelsize=13)


# plot in the fourth subplot
sns.heatmap(data=df_4k, ax=axs[1, 1], cmap='crest',annot=True, fmt=".1f", cbar=False, linewidths=.5, annot_kws={"fontsize": 12})
axs[1, 1].set_title('Dim = 4096', size=15)
axs[1, 1].set_yticks([])
axs[1, 1].set_ylabel('')
axs[1, 1].set_xlabel('Reference', size=18)
axs[1, 1].tick_params(axis='x', which='both', length=0, labelsize=13)

custom = {"axes.edgecolor": "black", "grid.linestyle": "dashed", "grid.color": "black"}
sns.set_style("whitegrid", rc = custom)
fig.set_figwidth(8)
fig.set_figheight(2.5)

# adjust the spacing between subplots
plt.subplots_adjust(left=0.1, bottom=0.1, right=0.9, top=0.9, wspace=0.05, hspace=0.35)
fig.savefig('/nfshome/yuz057/2023MICRO-MPCMXU/Figures/KNN.pdf',bbox_inches='tight')
# display the plot
plt.show()
```
